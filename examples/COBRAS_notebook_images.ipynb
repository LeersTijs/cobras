```python
## Loading necessary modules 
import os
import time
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython import display

import numpy as np
from dtaidistance import dtw

from cobras_ts.cobras_kmeans import COBRAS_kmeans
from cobras_ts.querier.notebookquerier_images import NotebookQuerierImages
from cobras_ts.cli.image_to_feature_vec import convert_img_to_feature_vec

budget = 100
```

```python
## Loading data, and constructing feature vectors using pre-trained network

images_path = '/home/jonas/Downloads/images'
network_path = '/home/jonas/Projects/cobras/cobras_ts/classify_image_graph_def.pb'
fns, features = convert_img_to_feature_vec(images_path, network_path=network_path)
print(len(fns))
```

```python
## Starting the clustering process
## Queried pairs will be plotted in the notebook
## A must-link is given by entering 'Y' in the notebook prompt, a cannot-link by entering 'n'.
## Each time an intermediate clustering is shown, you can decide to stop clustering (by entering 'n' in the prompt),\
## or continue answering queries (by entering  'Y' in the prompt).
## In the intermediate results, each row shows 5 random images from each cluster.
clusterer = COBRAS_kmeans(features, NotebookQuerierImages(fns), budget)
clustering, intermediate_clusterings, runtimes, ml, cl = clusterer.cluster()
```
